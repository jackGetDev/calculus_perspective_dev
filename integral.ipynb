```python
import matplotlib.pyplot as plt

def trapezoidal_integral(times, positions):
    total_distance = 0.0
    for i in range(1, len(times)):
        interval = times[i] - times[i - 1]
        average_velocity = (positions[i] + positions[i - 1]) / 2
        total_distance += average_velocity * interval

    return total_distance

time_data = [0, 1, 2, 3, 4]
position_data = [0, 3, 6, 7, 9]

total_distance_trapezium = trapezoidal_integral(time_data, position_data)
print("Total distance (Trapezoidal):", total_distance_trapezium, "meters")

# Visualisasi data pergerakan pemain
plt.figure(figsize=(8, 5))
plt.plot(time_data, position_data, marker='o', color='b')
plt.xlabel('Time (seconds)')
plt.ylabel('Position (meters)')
plt.title('Player Movement During the Match')
plt.grid(True)
plt.show()
```
